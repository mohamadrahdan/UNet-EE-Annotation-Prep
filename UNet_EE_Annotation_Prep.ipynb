```python
import ee
import folium # Import folium for interactive map display

ee.Authenticate()
ee.Initialize(project='poised-ceiling-461507-d5')

# Define the 'komeh' geometry
komeh = ee.FeatureCollection('projects/poised-ceiling-461507-d5/assets/areaKomeh')

# map visualization
centroid = komeh.geometry().centroid().coordinates().getInfo()
lon, lat = centroid[0], centroid[1]
map_object = folium.Map(location=[lat, lon], zoom_start=10)

# Add the 'komeh' boundary to the Folium map
map_object.add_child(
    folium.GeoJson(
        komeh.getInfo(),
        name='Komeh Boundary'
    )
)

map_object.add_child(folium.LayerControl())
display(map_object)

# Filter the Sentinel-2 image collection
images = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")\
    .filterBounds(komeh)\
    .filterDate('2015-01-01', '2024-06-09')\
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))
#print(images)

# To view properties of the image collection (e.g., size, first image):
print(f"Number of images in collection: {images.size().getInfo()}")
print(f"First image in collection: {images.first().getInfo()}")
```
